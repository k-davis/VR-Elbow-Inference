```python
import tensorflow as tf
from tensorflow.keras.callbacks import TensorBoard
import time
import sys
```

```python
sys.path.append('..')
from KerasUtils import NonTimeSeriesFormatter, Metrics
```

```python
mocap_file = "80_postproc_full.json"
input_labels = ["lhumerus", 'lwrist']
output_label = "lradius"

x_train, y_train, x_test, y_test = NonTimeSeriesFormatter.get_data(mocap_file,
                             input_labels, output_label)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(8, activation='linear', input_shape=(3*len(input_labels),)))
model.add(tf.keras.layers.Dense(8, activation='linear'))
model.add(tf.keras.layers.Dense(3, activation='linear'))
```

```python
model.compile(optimizer='adam',
             loss=Metrics.mse_between_points,
             metrics=[Metrics.mse_between_points, Metrics.mean_diff_error])
```

```python
tensorboard = TensorBoard(log_dir='logs\SimpleModel\{}'.format(time.strftime("%d-%m-%y_%H.%M.%S")))
model.fit(x_train, y_train, batch_size=64, epochs=20, callbacks=[tensorboard])
```

```python
val_loss, val_mse_b_p, val_mean_diff = model.evaluate(x_test, y_test)
```
