```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
import json
import os
```

```python
# process data
x_feature_labels = ["lclavicle", "thorax"]
y_feature_labels = ["rclavicle"]
input_data = {"x": [], "y": []}

def readFolder(folder_path):
    for file in os.listdir(folder_path):
        if file.endswith(".json"):
            readFile(folder_path + "\\" + file)

def readFile(file_path):
    full_mocap = json.load(open(file_path))
    
    for frame in full_mocap.values():
        frame_x_data = []
        for label in x_feature_labels:
            frame_x_data.extend(frame[label]["coordinate"])
        input_data["x"].append(frame_x_data)
            
        frame_y_data = []
        for label in y_feature_labels:
            frame_y_data.extend(frame[label]["coordinate"])
        input_data["y"].append(frame_y_data)

def normalize():
    x_min = min([frame_coords for frame_coords in input_data["x"]])
    

def splitIntoTrainingTesting():
    TRAINING_PERCENT = .70
    TRAINING_COUNT = int(TRAINING_PERCENT * len(input_data["x"]))
    x_train = input_data["x"][:TRAINING_COUNT]
    y_train = input_data["y"][:TRAINING_COUNT]
    x_test = input_data["x"][TRAINING_COUNT:]
    y_test = input_data["y"][TRAINING_COUNT:]
    return x_train, y_train, x_test, y_test

def shuffle():
    pass
    
```

```python
#readFolder("E:\Desktop\github\VR-Elbow-Inference\mocap_data\80")
readFolder('E:\Desktop\THESIS\subject_data\80')
x_train, y_train, x_test, y_test = splitIntoTrainingTesting()
```

```python
model = tf.keras.models.Sequential()
#model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(16, activation='relu', input_shape=(6,)))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(3, activation='relu'))

```

```python
#training params
# these params have no meaning to me
model.compile(optimizer='adam',
             loss='mean_squared_error',
             metrics=['mean_squared_error'])
```

```python
model.fit(x_train, y_train, epochs=3)
```

```python
val_loss, val_acc = model.evaluate(x_test, y_test)
print(val_loss, val_acc)
```
